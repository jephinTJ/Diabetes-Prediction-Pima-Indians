# DIABETES PREDICTION SYSTEM

```python
# Import required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Libraries for prediction
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("D:\GitHub\Diabetes-Prediction-Pima-Indians\source data\diabetes.csv")
```

```python
data
```

```python
# checking for missing values
sns.heatmap(data.isnull())
```

```python
# co-relation matrix
correlation=data.corr()
print(correlation)
```

```python
plt.title("Correlation Matrix")
sns.heatmap(correlation, annot=False)
# darker the color lesser correlation between 2 columns
```

```python
# traindata & testdata splitting
X=data.drop("Outcome",axis=1)
Y=data['Outcome']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
# training the model
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
# making prediction
prediction=model.predict(X_test)
print(prediction)
```

```python
# accuracy check
accuracy=accuracy_score(prediction,Y_test)
print(accuracy)
```
